## Correlation

```python
# Correlation Matrix Heatmap
corrmat = df.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, square=True)
```

```python
# Top 10 Heatmap
k = 10 #number of variables for heatmap
cols = corrmat.nlargest(k, 'price')['price'].index
cm = np.corrcoef(df[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, 
                 yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```

```python
most_corr = pd.DataFrame(cols)
most_corr.columns = ['Most Correlated Features']
most_corr
```

## Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
model = LinearRegression()
```

```python
# fit the model using these variables
features = ['sqft_living', 'grade', 'bathrooms', 'view', 'sqft_basement', 'bedrooms', 'floors', 'condition', 'yr_built']
```

```python
# split the dataset into 80% training and 20% testing
x_train , x_test , y_train , y_test = train_test_split(df[features], df['price'] , test_size = 0.20, random_state=2)
```

```python
model.fit(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
# y-intercept
model.intercept_
```

```python
# coefficients
model.coef_
```
